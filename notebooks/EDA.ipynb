```python
#Importing Base Libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Importing the processed dataset
df = pd.read_csv(r'C:\Users\Deep Joshi\Desktop\Data Analysis\Machine Learning\Income Prediction\data\processed\CensusDataProcessed.csv')
df.head()
```

```python
sns.countplot(x='Income',data=df,hue = 'Gender')
# We can clearly differentiate the relationship between gender and income class.
# The count of males having income >50K is much more than that of females (The ratio is close 1:7)
```

```python
sns.countplot(x='Income',data=df,hue = 'Education Number')
# Education Level plays an important role in determining income category.
# Individuals having 13 years of formal education have the most count under >50K category. It is then followed by 9 and 10 years of education.
# This again depends on other attributes like occupation, age, work class,etc.
```

```python
sns.countplot(x='Income',data=df,hue = 'Race')
# According to our analysis, individuals identified to be coming from white race have the highest count under both the categories.
# It is then followed by individuals coming from black race, but the difference is quite significant
```

```python
sns.countplot(x='Income',data=df,hue ='Work Class')
# Private sector of work class have maximum number of individuals. Out of which 70% of individuals have income >50K 
```

```python
sns.countplot(x='Income',data=df,hue = 'Marital Status')
# The relationship between income and marital status is interesting. In <=50K class, independent individuals have higher count than married individuals.
# Where as, in the >50K class, married individuals have more count than independent people. 
```

```python

sns.jointplot(x=df['Age'],y=df['Hours/Week'],data=df,kind='hex',cmap='hot')
# From the below jointplot, it can be identified that most of the individuals across all the age groups work aroung 40 hours/week
```

```python
sns.scatterplot('Capital Gain','Capital Loss', data= df)
# From the scatter plot between capital gain and capital loss, we can gain various insights:
# 1. Capital Gain and Capital Loss can be 0 - This indicate no capital change
# 2. If Capital Gain is 0, Capital Loss can be 0 or above 0 and vice versa. 
# 3. We can find the capital change using both the attributes i.e. Capital Gain - Capital Loss
```
